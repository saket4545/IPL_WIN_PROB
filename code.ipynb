```python
import pandas as pd
import numpy as np
```

```python
matches= pd.read_csv('IPL_Matches_Result_2008_2022.csv')
matches.head(5)
```

```python
print('numpy',np.__version__)
print('pandas',pd.__version__)
print('sklearn',scikit-learn.__version__)
```

```python
delivery= pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head(5)
```

```python
delivery = delivery.merge(matches[['ID', 'Team1', 'Team2']], on='ID', how='left')


delivery['BowlingTeam'] = delivery.apply(
    lambda row: row['Team1'] if row['BattingTeam'] == row['Team2'] else row['Team2'], axis=1
)
```

```python
delivery.head()
```

```python
total_score=delivery.groupby(['ID','innings'])['total_run'].sum().reset_index()
total_score=total_score[total_score['innings']==1]
total_score
```

```python
#target is always +1 from total runs
total_score['total_run'] = total_score['total_run'] + 1
total_score.head(5)
```

```python
match_df=matches.merge(total_score[['ID','total_run']],left_on='ID',right_on='ID')
match_df.head(5)
```

```python
match_df['Team1'].unique()
```

```python
teams=['Rajasthan Royals',
       'Royal Challengers Bangalore',
       'Sunrisers Hyderabad',
       'Delhi Capitals',
       'Chennai Super Kings',
       'Gujarat Titans',
       'Lucknow Super Giants',
       'Kolkata Knight Riders',
       'Punjab Kings',
       'Mumbai Indians'
]
```

```python
match_df['Team1']= match_df['Team1'].str.replace("Kings XI Punjab","Punjab Kings")
match_df['Team2']= match_df['Team2'].str.replace("Kings XI Punjab","Punjab Kings")

match_df['Team1']= match_df['Team1'].str.replace("Delhi Daredevils","Delhi Capitals")
match_df['Team2']= match_df['Team2'].str.replace("Delhi Daredevils","Delhi Capitals")

match_df['Team1']= match_df['Team1'].str.replace("Deccan Chargers","Sunrisers Hyderabad")
match_df['Team2']= match_df['Team2'].str.replace("Deccan Chargers","Sunrisers Hyderabad")

match_df['Team1']= match_df['Team1'].str.replace("Gujarat Lions","Gujarat Titans")
match_df['Team2']= match_df['Team2'].str.replace("Gujarat Lions","Gujarat Titans")

```

```python
match_df= match_df[match_df['Team1'].isin(teams)]
match_df= match_df[match_df['Team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
match_df=match_df[['ID','City','WinningTeam','total_run']]
```

```python
delivery_df=match_df.merge(delivery,on='ID')
delivery_df.head()
```

```python
delivery_df=delivery_df[delivery_df['innings']==2]
delivery_df.shape
```

```python
delivery_df['current_score']=delivery_df.groupby('ID')['total_run_y'].cumsum()
delivery_df.head(10)
```

```python
delivery_df['runs_left']=delivery_df['total_run_x']-delivery_df['current_score']
delivery_df.head(10)
```

```python
delivery_df['ballnumber_corrected'] = (delivery_df['overs'] * 6) + delivery_df['ballnumber']
delivery_df.head(10)
```

```python
total_overs = 20
total_balls_in_match = total_overs * 6

delivery_df['balls_left']=total_balls_in_match - delivery_df['ballnumber_corrected']
delivery_df.head(12)
```

```python
delivery_df['player_out'].unique()
```

```python
delivery_df['overs_faced'] = delivery_df['ballnumber_corrected'] / 6
delivery_df['CRR'] = delivery_df['current_score'] / delivery_df['overs_faced']
delivery_df.head(5)
```

```python
delivery_df['req_runrate'] = (delivery_df['runs_left'] / delivery_df['balls_left']) * 6
delivery_df.head(5)
```

```python
def result (row):
    return 1 if row['BattingTeam']==row['WinningTeam'] else 0
```

```python
delivery_df['result']= delivery_df.apply(result,axis=1)
delivery_df.head(6)
```

```python
delivery_df.columns
```

```python
final_df=delivery_df[['BattingTeam','BowlingTeam','City',"batter","bowler",'runs_left','balls_left','total_run_x','CRR','req_runrate','result']]
```

```python
final_df['req_runrate'] = final_df['req_runrate'].round(2)
final_df['CRR'] = final_df['CRR'].round(2)
```

```python
final_df.head()
final_df.dropna(inplace=True)
```

```python
final_df.sample(final_df.shape[0])
```

```python
final_df=final_df[final_df['balls_left'] !=0]
```

```python
print(type(final_df))
```

```python
X=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
```

```python
X['req_runrate'].tail()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(drop='first',handle_unknown='ignore'),['BattingTeam','BowlingTeam','City','batter','bowler'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),  
                   
    ('step2', LogisticRegression(solver='liblinear'))
    ])
```

```python
X_train.describe()
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)


print(f"Accuracy: {accuracy}")
```

```python
pipe.predict_proba(X_test)[45]
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print("\nClassification Report:\n")
print(report)
```
